The resnet50 has 3 layer groups.
The last one (3rd) is the fully connected fine-tuning layer.
The 2nd one has 4 layer subgroups:
 
 1. the 1st has 6 bottleneck layers
 2. the 2nd has 3 bottleneck layers
 3. the 3rd is an AdaptiveConcatPool2d
 4. the 4th is a Flatten layer
 
So we're going to unfreeze 2nd to 4th layers

```python
def set_trainable_layers(learner, layer_subgroup, from_layer, to_layer):
    learner.unfreeze()
    for layer in range(from_layer-1, to_layer):
        learner.get_layer_groups()[layer_subgroup-1][layer].trainable = True
#     return learner
```

```python
set_trainable_layers(learner, 1, 2,4)
```

```python
def rand_by_mask(mask, n=4): return np.random.choice(np.where(mask)[0], n, replace=False)
def rand_by_correct(is_correct, n=4): return rand_by_mask((preds == data.val_y)==is_correct, n)
```

```python
def plot_val_with_title(idxs, title, rows=1):
    imgs = np.stack([data.val_ds[x][0] for x in idxs])
    title_probs = [probs[x] for x in idxs]
    print(title)
    return plots(data.val_ds.denorm(imgs), rows=rows, titles=title_probs)
```

```python
def plots(ims, figsize=(12,6), rows=1, titles=None):
    f = plt.figure(figsize=figsize)
    for i in range(len(ims)):
        sp = f.add_subplot(rows, len(ims)//rows, i+1)
        sp.axis('Off')
        if titles is not None: sp.set_title(titles[i], fontsize=16)
        plt.imshow(ims[i])
```

```python
def load_img_id(ds, idx): return np.array(PIL.Image.open(PATH+ds.fnames[idx]))

def plot_val_with_title(idxs, title, rows=1):
    imgs = [load_img_id(data.val_ds,x) for x in idxs]
    title_probs = [probs[x] for x in idxs]
    print(title)
    return plots(imgs, rows=rows, titles=title_probs, figsize=(16,8))
```

```python
# 1. A few correct labels at random
plot_val_with_title(rand_by_correct(True, n=None), "Correctly classified")
```

```python
plot_val_with_title(rand_by_correct(False, n=None), "InCorrectly classified", rows=2)
```
