```python
def evaluate(learner, aug=True):
    mean, y = None, None
    if (aug):
        log_preds,y = learner.TTA()
        mean = np.mean(np.exp(log_preds), 0)
    else:
        mean, y = learner.predict_with_targs()
    print(accuracy(mean,y), metrics.log_loss(y, mean))
    preds = np.argmax(mean, axis=1)
    probs = np.exp(mean[:,1])
    from sklearn.metrics import confusion_matrix
    cm = confusion_matrix(y, preds)
    plot_confusion_matrix(cm, data.classes)
```

# Initialize Model - image size 224x224

```python
arch=resnet50
bs=32
sz=224
```

```python
def get_seed(n, is_r=False):
    return nt(time())+n if is_r else 42

def get_val_idxs(n, val_pct=0.2, random=False):        
    v = [get_cv_idxs(n, val_pct=val_pct, seed=get_seed(i*n))+i*n for i in range(4)]
    return np.hstack(v)
```

```python
def get_data(arch, sz, bs, val_idxs):
    transf = transforms_basic + [RandomFlipXY(), RandomDihedralXY()]
    tfms = tfms_from_model(arch, sz, aug_tfms=transf, max_zoom=1.1)
    csv_fname = PATH_CSV+'microscopy_ground_truth.csv'
   
    data = ImageClassifierData.from_csv(PATH_CSV, 'train', csv_fname, bs,tfms,val_idxs)
    return data
```

```python
val_idxs = get_val_idxs(99)
val_idxs[:3], val_idxs[100:103]
```

```python
data = get_data(arch, sz, bs, val_idxs)
```

```python
learner = ConvLearner.pretrained(arch, data, precompute=True, ps=0.4, )
learner.lr_find()
learner.sched.plot()
```
